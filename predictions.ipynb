```python
from sklearn.preprocessing import OneHotEncoder
```

```python
object_columns = df_train.select_dtypes(include='object').columns
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoder_array = encoder.fit_transform(df_train[object_columns])
```

```python
encoder_array
```

```python
encoded_df = pd.DataFrame(encoder_array, columns = encoder.get_feature_names_out(object_columns))
```

```python
encoded_df
```

```python
encoded_train_df = pd.concat([df_train.drop(columns=object_columns), encoded_df], axis=1)
```

```python
encoded_train_df
```

```python
set(encoded_df.columns) - set(pd.get_dummies(df_train.select_dtypes(include='object')).columns)
```

```python
encoder.categories_
```

```python
nan_count = encoded_train_df.isna().sum()
nan_count[nan_count > 0]
```

```python
##Use mean to fill NaN integers and float
```

```python
encoded_train_df["LotFrontage"] = encoded_train_df["LotFrontage"].fillna(encoded_train_df["LotFrontage"].mean())
```

```python
encoded_train_df["MasVnrArea"] = encoded_train_df["MasVnrArea"].fillna(encoded_train_df["MasVnrArea"].mean())
```

```python
encoded_train_df["GarageYrBlt"] = encoded_train_df["GarageYrBlt"].fillna(encoded_train_df["GarageYrBlt"].mode()[0])
```

```python
X_train = encoded_train_df.drop(columns=['SalePrice','Id'])
Y_train = encoded_train_df["SalePrice"]
```

```python
fit = bestFeatures.fit(X_train, Y_train)
```

```python
fit
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X_train.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
```

```python
featureScores
```

```python
featureScores.columns = ['FeatureNames','Score'] 
```

```python
featureScores
```

```python
featureScores.nlargest(20,'Score').sort_values("Score", ascending=False)  #print 10 best features
```
